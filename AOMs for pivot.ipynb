## R script to generate AOM origins around a pivot

#### Clear the environment.

```r
rm(list=ls()) 
```

#### Import data.table and ggplot2 libraries. 

```r
library(data.table) 
library(ggplot2) 
```

#### calculate theta

```r
theta <- seq(0, 2*pi, (2*pi)/269)
```

#### create a series of radii

```r
radii <- seq(25,130,2) 
```

#### create a series of theta values for each value in radii

```r
theta_all <- rep(theta, length(radii))
```

#### create an equally long series of radii values for calculation of (x,y) x and y

```r
radii_all <- sort(rep(radii, length(theta)))
```

#### make a data.table of vectors created

```r
DT <- data.table(radii=radii_all, theta = theta_all)
```

#### calculate values for x for coordinates used as AOM origin

```r
DT[, x := round((-11343682.475+radii*cos(theta)),3)]
```

#### calculate values for y for coordinates used as AOM orign

```r
DT[, y := round((3974521.392+radii*sin(theta)),3)]
```

#### Plot the ordered pairs we just created.

```r
p1 <- ggplot(DT)+
        geom_point(aes(x=x,y=y), size = .1)
p1
```

#### Now we can generate points for polygons. First a shape ID vector.

```r
shape_id <- sort(rep(seq(1,length(DT$x),1), 12))
```

#### A new theta will be used so that there will only be 12 points in each polygon.

```r
theta_p <- seq(0, 2*pi, (2*pi)/11)
```

#### Create a series of theta values for each origin that we created above

```r
theta_poly <- rep(theta_p, length(DT$x))
```

#### Create a series of radii values to be used to in the calculation of x and y.

```r
radii_poly <- rep(1, length(theta_poly))
```
