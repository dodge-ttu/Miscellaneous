### Import required libraries.

```python
import pandas as pd
from matplotlib import pyplot as plt # this is a plotting libarary
```

###  Create an object with the file path to the desired raw data .csv.

```python
filename = 'Data_Atinder.csv'
```

### Read the .csv file and use the "header" argument to identify multi-line header in raw data.

```python
data = pd.read_csv(filename, header=[0,1,2])
```

### Create an object of column names from the mult-line header.

```python
column_names = [' '.join(i) for i in data.columns.values]
```

### Assign the column object we just created to the column name variable to create neat column headers.

```python
data.columns = column_names
```

### Convert the datetime object to a datetime class that Python understands.

```python
data['EM-4th 817 records Measurement Time'] = pd.to_datetime(data['EM-4th 817 records Measurement Time'], format = '%m/%d/%Y %I:%M %p')
```

### Set the index of the data frame to the datetime column so that we can resample and aggregate the data.

```python
data = data.set_index(data['EM-4th 817 records Measurement Time'])
```

### Create new data frames with the desired datetime granularity.

```python
data_15min = data.resample('15MIN').mean()
data_30min = data.resample('30MIN').mean()
data_60min = data.resample('1H').mean()
```

### Plot pressure variables from a couple sensors.

```python
potential_cols = data_60min[['Port 2 MPS-6 Water Potential/Temp kPa Potential',
                            'Port 4 MPS-6 Water Potential/Temp kPa Potential']]

cols_potential = list(potential_cols)
plt.figure(figsize=(18,6.5))

for i in cols_potential:
    plt.plot(potential_cols[i][20:len(potential_cols)])

plt.legend(loc='lower left')
```

### Plot  VWC variable for several sensors

```python
VWC_cols = data_60min[['Port 1 5TE Moisture/Temp/EC m³/m³ VWC',
                       'Port 3 5TE Moisture/Temp/EC m³/m³ VWC',
                       'Port 5 5TE Moisture/Temp/EC m³/m³ VWC']]

cols_VWC = list(VWC_cols)
plt.figure(figsize=(18,6.5))

for i in cols_VWC:
    plt.plot(VWC_cols[i][20:len(VWC_cols)])

plt.legend(loc='lower left')
```

### Plot multiple variables for a single sensor.

```python
Port1 = data_60min[['Port 1 5TE Moisture/Temp/EC m³/m³ VWC',
                     'Port 1 5TE Moisture/Temp/EC °C Temp',
                     'Port 1 5TE Moisture/Temp/EC mS/cm EC Bulk']]

cols = list(Port1)
plt.figure(figsize=(18,6.5))

for i in cols:
    plt.plot(Port1[i][20:len(Port1)])

plt.legend(loc='upper right')
```

### Uh oh! Our parameter values are too far apart, let's scale our data to see relationships.

#### Here we see that we can take the standard deviation of the data for each variable.

```python
data_60min.std()
```

#### We can also take the mean of the data for each variable.

```python
data_60min[0:10].mean()
```

#### To scale the data we find the difference between a value and its mean and divide by the standard deviation.

```python
data_60min_scaled = (data_60min - data_60min.mean()) / data_60min.std()
```

### Plot scaled data.

#### Our scaled data represents the number of standard deviations from the mean.  We can see how changes in one variable relate to changes in another becuase the scaled data has values that can be plotted on a common axes.

```python
Port1 = data_60min_scaled[['Port 1 5TE Moisture/Temp/EC m³/m³ VWC',
                           'Port 1 5TE Moisture/Temp/EC °C Temp',
                           'Port 1 5TE Moisture/Temp/EC mS/cm EC Bulk']]

cols_port1 = list(Port1)
plt.figure(figsize=(18,10))

for i in cols_port1:
    plt.plot(Port1[i][20:len(Port1)])
    
plt.legend(prop={'size':12})
```

#### Plot all variables in the scaled data set just for fun!

```python
cols_60min_scaled = list(data_60min_scaled)
plt.figure(figsize=(18,10))

for i in cols_60min_scaled:
    plt.plot(data_60min_scaled[i][20:len(data_60min_scaled)])
    
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, prop={'size':16})
```

### Now to save a plot we need to add a line at the end of our plotting code. 

```python
Port1 = data_60min_scaled[['Port 1 5TE Moisture/Temp/EC m³/m³ VWC',
                           'Port 1 5TE Moisture/Temp/EC °C Temp',
                           'Port 1 5TE Moisture/Temp/EC mS/cm EC Bulk']]

cols_port1 = list(Port1)
plt.figure(figsize=(18,10))

for i in cols_port1:
    plt.plot(Port1[i][20:len(Port1)])
    
plt.legend(prop={'size':12})
plt.savefig('new_plot.pdf') # this line saves the plot to the given filename
```
