<h1 style="text-align:center;padding:0">What would random bad seed look like in a field?</h1>

A discussion arose as to how likely it would be to have four or five or six bad seeds planted in a row if the seeds were randomly mixed in the planter. The speculation was that in a large population one might find gaps of several plants even though the likelihood of might be low if ony a small percentage of seed was bad.

Of course this can be shown mathematically with ease but the visual representation is helpful when looking at a cotton field one might suspect has been affected by a percentage of bad seed.

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from io import StringIO
```

### Method One:

```python
n = 100
s = 100

x=[i for i in range(1,n+1,1)]*s
y=sorted([i for i in range(1,n+1,1)]*s)
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,16))
ax.plot(x,y, 'go', markersize=3)
ax.set_title('100 Percent Good Seed')
```

```python
# enter a bad seed value out of n * s 
bad_seed_number = 50

b = [i for i in np.random.randint(0,len(x)-1, bad_seed_number)]

for i in b:
     x[i] = np.nan
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,16))
ax.plot(x,y, 'go', markersize=3)
#ax.set_title('90 Percent Good Seed')
```

```python
n = 100
s = 100

bad_seed_percent = .01
good_seed_percent = 1 - bad_seed_percent

bad_seed_num = int(n * bad_seed_percent)
good_seed_num = int(n * good_seed_percent)

good_id = [1 for i in range(good_seed_num)]
bad_id = [0 for i in range(bad_seed_num)]

good_bad_id = good_id + bad_id
good_bad_id = good_bad_id * s
np.random.shuffle(good_bad_id)

x=[i for i in range(1,n+1,1)]*s
y=sorted([i for i in range(1,n+1,1)]*s)

len_x = len(x)

dict = {
    'x':x,
    'y':y,
    'good_bad': good_bad_id
}

print(len(good_bad_id))
print(len(x))
print(len(y))

df = pd.DataFrame(dict)

fig, ax = plt.subplots(1,1, figsize=(16,16))
ax.scatter(df.x, df.y, c=df.good_bad, cmap='winter', s=12)
```
