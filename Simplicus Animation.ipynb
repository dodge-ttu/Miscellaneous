```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math

from matplotlib import animation, rc
from IPython.display import HTML
```

```python
def circle_xy(theta, origin_x, origin_y, radius):
    # origin at (h,k) with radius (r)
    
    # x = radius * np.cos(deg) + h
    # y = radius * np.sin(deg) + k
    x = radius * np.cos(theta) + origin_x 
    y = radius * np.sin(theta) + origin_y
       
    return x,y
```

```python
def line_seg_coord(x,y):

    x,y = circle_xy(theta, -1, 3, 2)

    point_ones = [(x,y) for (x,y) in zip(x,y)]
    point_twos = [point_ones[i] for i in range(-1, len(point_ones),1)]

    points = [(p_one, p_two) for (p_one, p_two) in zip(point_ones, point_twos)]
    
    return points
```

```python
theta = [i for i in np.linspace(0, 2*np.pi, 10)]

x,y = circle_xy(theta, -1, 3, 2)

points = line_seg_coord(x,y)

fig, ax = plt.subplots(1,1, figsize=(10,10))

for (p_one, p_two) in points:
    x1,y1,x2,y2 = p_one[0],p_one[1],p_two[0],p_two[1]
    
    x = [x1,x2]
    y = [y1,y2] 
    
    ax.plot(x,y)
```

```python
# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return (line,)

# animation function. This is called sequentially
def animate(i):
    
    # origin at (h,k) with radius (r)
    h = -1
    k = +3
    radius = 1
    theta = [i for i in np.linspace(0, 2*np.pi, 360)]
    
    # x = radius * np.cos(deg) + h
    # y = radius * np.sin(deg) + k
    x = [radius * np.cos(i) + h for i in theta]
    y = [radius * np.sin(i) + k for i in theta]
    
    x = x * 4
    y = y * 4
    
    x = x[i:]
    y = y[i:]
    
    line.set_data(x, y)
    return (line,)

fig, ax = plt.subplots(1,1, figsize=(7,7))

line, = ax.plot([], [], lw=1)

ax.axis('off')
            
ax.set_xlim(( -2, 2))
ax.set_ylim(( -2, 2))

# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=1000, interval=30, blit=True)
HTML(anim.to_html5_video())
#anim.save('basic_animation.mp4', fps=30)
```
